```python
import json

scoore = []
size = []

for i in range(1, 100+1):
    file_name = f"scoore_{i}_.json"
    try:
    
            with open(file_name , "r") as file:
                data = json.load(file)
                length = len(data)
                
                scoore.append(i)
                size.append(length)

                print(f"Scoore: {i} --|-- Size: {length}")

    except FileNotFoundError:
            continue
        
```

<h1>I have tried to understand the data</h1>

```python
import matplotlib.pyplot as plt

plt.plot(scoore , size)
plt.xlabel("Scoore")
plt.ylabel("Size")
plt.title("Scoore vs Size")
plt.grid()
plt.savefig("scoore_vs_size.png")
plt.show()
```

```python
plt.bar(scoore , size)
plt.xlabel("Scoore")
plt.ylabel("Size")
plt.title("Scoore vs Size Bar")
plt.grid()
plt.savefig("scoore_vs_size_bar_.png")
plt.show()
```

```python
plt.pie(size )
plt.title("Scoore vs Size Pie")
plt.savefig("scoore_vs_size_pie_.png")
plt.show()
```

```python
total_size = sum(size)
print(f"Total Size: {total_size}")

def calculate_the_per(lenth):
    return (lenth*100 )/ total_size

def calculate_the_per_int(lenth):
    return (lenth*100 )// total_size
```

```python
percentages = [calculate_the_per(l) for l in size]
int_percentages = [calculate_the_per_int(l) for l in size]

print(int_percentages)
print(sum(int_percentages))

print(percentages)
print(sum(percentages))

def filter_int_percentage_(datas):
    scoore = []
    percentages = []
    
    for data in datas:
        if data > 0:
            scoore.append(datas.index(data)+1)
            percentages.append(data)
    return scoore , percentages

filtered_scoore , filtered_percentages = filter_int_percentage_(int_percentages)

print(filtered_scoore)
print(filtered_percentages)
print(sum(filtered_percentages))
    
```

```python
plt.bar(filtered_scoore , filtered_percentages)
plt.xlabel("Scoore")
plt.ylabel("Size")
plt.title("Scoore vs Size in int_percent")
plt.grid()
plt.savefig("scoore_vs_size_bar_int_pecent.png")
plt.show()
```

```python
plt.pie(filtered_percentages )
plt.title("Scoore vs Size in int_percent")
plt.savefig("scoore_vs_size_pie_int_percent.png")
plt.show()
```
