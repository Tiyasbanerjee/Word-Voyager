```python
for i in range(4,22):
    file_names = f"words_len_{i}_.txt"
    print(f"Run for ---------- file_{i}")
   
    with open(file_names , "r") as f:
        words = f.read().splitlines()
    words = [word for word in words]
   
    length = len(words)
    
    last_index = 0
    list_of_tokens = []
  
    for j in range(100,length,100):
        tockened_list = words[last_index:j]
        list_of_tokens.append(tockened_list) #type: ignore
        last_index = j
    
    list_of_tokens.append(words[last_index:]) #type: ignore

    dict_to_save = {}
    
    
    token_num = 1
    for tokens in list_of_tokens:  #type:ignore
        save_token_name = f"file_{i}_token_{token_num}_.text"
        print(f"Run for ---------- {save_token_name}")
        token_num += 1
        with open(save_token_name , "w") as f:
            f.write(str(tokens))   #type: ignore



```
