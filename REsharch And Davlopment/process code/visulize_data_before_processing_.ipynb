```python
import os
dir_name = os.getcwd() + "/"


def count_lines(file_path:str):
        with open(file_path, mode='r') as file_:

           line_count = sum( 1 for line in file_ )
        return line_count

list_ = []

for i in range(1,26):
    file_path = dir_name + f"words_len_{i}_.txt"
    list_.append(count_lines(file_path))



```

<h1>Before starting i tried to understand the raw data</h1>

```python
print(list_)

list_of_names = [ str(p) for p in range(1,26)]
print(list_of_names)
```

<h1>We will plot a Graph to understand the distrubution</h1>

```python
import matplotlib.pyplot as plt

plt.bar(list_of_names , list_)

plt.savefig("bar_plot.png")
```

<h1>Here it is clearly a normal distrution</h1>

```python
result_colour = plt.pie(list_)

plt.savefig("pie_plot.png")
```

<h1>The pie chart here</h1>

```python

colors = [w.get_facecolor() for w in result_colour[0]]
```

```python
from matplotlib.patches import Rectangle
from matplotlib.colors import to_hex
```

```python
def detalils(num:int , total:int):
    return (num/total)*100


fig , ax = plt.subplots(figsize=(5.5,5.5))
ax.set_xlim(- 0.8, 0.8)  
ax.axis('off')

i=1
tatal = sum(list_)

for i, c in enumerate(colors):
    ax.plot( -0.6, i, 'o', color=c, markersize=10) 
    ax.text(-0.5 , i , f"---> || {i}'s percentage : {detalils(list_[i-1] ,tatal):.2f} % ||")

ax.invert_yaxis()
fig.savefig("pie_plot_details.png")

```
