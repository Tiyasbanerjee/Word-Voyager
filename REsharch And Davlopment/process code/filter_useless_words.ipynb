```python
import json 

file_name = 'data_1_to_10.json'

filtured = {}

with open(file_name, 'r') as file:
    data = json.load(file)

print(len(data))

unw = ['একটি',
        'অপরিচিত শব্দ'
]

for key, values_list in data.items():


    translation = values_list[0] 


    is_unwanted = any(keyword in translation for keyword in unw)

    if not is_unwanted:
        filtured[key] = translation



print(len(filtured))

with open('level_1.json', 'w') as outfile:
    json.dump(filtured, outfile, indent=4 , ensure_ascii=False)
```
